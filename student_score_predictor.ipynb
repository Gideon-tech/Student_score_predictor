```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Study Hours vs Scores
data = {
    "Hours": [1, 2, 3, 4.5, 5, 6.5, 7, 8, 9, 10, 11],
    "Scores": [10, 20, 30, 45, 50, 65, 70, 80, 90, 95, 100]
}

df = pd.DataFrame(data)
print(df)
```

```python
plt.scatter(df["Hours"], df["Scores"], color="blue")
plt.title("Study Hours vs Score")
plt.xlabel("Hours Studied")
plt.ylabel("Score")
plt.grid(True)
plt.show()
```

```python
X = df[["Hours"]]  # Features (input)
y = df["Scores"]   # Target (output)

model = LinearRegression()
model.fit(X, y)
```

```python
# Predict score for 6 hours
predicted_score = model.predict([[6]])
print(f"If a student studies 6 hours, they will likely score: {predicted_score[0]:.2f}")
```

```python
line = model.coef_ * X + model.intercept_
plt.scatter(X, y, color='blue')
plt.plot(X, line, color='red')
plt.title("Regression Line")
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.show()
```
